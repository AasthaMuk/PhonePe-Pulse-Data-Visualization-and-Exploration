Data Extraction for Aggregated transactions

```python
path = data_dir+"/data/aggregated/transaction/country/india/state/"
aggre_state = os.listdir(path)

aggr_transactions = {'State': [], 
                 'Year': [], 
                 'Quarter': [], 
                 'Transaction_type': [], 
                 'Transaction_count': [],
                 'Transaction_amount': []
                }

for state in aggre_state:
    # to access each state folder say "assam" folder inside "data/aggregated/transaction/country/india/state"
    cur_state = path+state+"/" 
    aggre_year = os.listdir(cur_state)

    for year in aggre_year:
        # to access each year folder say "2018" folder inside "data/aggregated/transaction/country/india/state/assam/2018"
        cur_year = cur_state+year+"/"
        aggre_files = os.listdir(cur_year)

        for file in aggre_files:
            # to access each json file say "1.json" folder inside "data/aggregated/transaction/country/india/state/assam/2018/1.json"
            cur_file = cur_year+file
            with open(cur_file,"r") as f:
                info = json.load(f)

            for item in info['data']['transactionData']:
                aggr_transactions['Transaction_type'].append(item['name'])
                aggr_transactions['Transaction_count'].append(item['paymentInstruments'][0]['count'])
                aggr_transactions['Transaction_amount'].append(item['paymentInstruments'][0]['amount'])
                aggr_transactions['State'].append(state)
                aggr_transactions['Year'].append(year)
                aggr_transactions['Quarter'].append(int(file.strip(".json")))

df_aggr_trans = pd.DataFrame(aggr_transactions)
```

Data Extraction for Aggregated users

```python
path = data_dir+"/data/aggregated/user/country/india/state/"
aggre_state = os.listdir(path)

aggr_users = {'State': [], 
              'Year': [], 
              'Quarter': [], 
              'Brands': [], 
              'Count': [],
              'Percentage': []
            }

for state in aggre_state:
    # to access each state folder say "assam" folder inside "data/aggregated/transaction/country/india/state"
    cur_state = path+state+"/" 
    aggre_year = os.listdir(cur_state)

    for year in aggre_year:
        # to access each year folder say "2018" folder inside "data/aggregated/transaction/country/india/state/assam/2018"
        cur_year = cur_state+year+"/"
        aggre_files = os.listdir(cur_year)

        for file in aggre_files:
            # to access each json file say "1.json" folder inside "data/aggregated/transaction/country/india/state/assam/2018/1.json"
            cur_file = cur_year+file
            with open(cur_file,"r") as f:
                info = json.load(f)

            # print(info['data']['usersByDevice'])
            try:
                for item in info['data']['usersByDevice']:
                    aggr_users['Brands'].append(item['brand'])
                    aggr_users['Count'].append(item['count'])
                    aggr_users['Percentage'].append(item['percentage'])
                    aggr_users['State'].append(state)
                    aggr_users['Year'].append(year)
                    aggr_users['Quarter'].append(int(file.strip(".json")))
            except:
                pass

df_aggr_users = pd.DataFrame(aggr_users)
```

Data Extraction for Map transactions

```python
path = data_dir+"/data/map/transaction/hover/country/india/state/"
map_state = os.listdir(path)

map_transactions = {'State': [], 
                     'Year': [], 
                     'Quarter': [], 
                     'District': [], 
                     'Count': [],
                     'Amount': []
                    }

for state in map_state:
    # to access each state folder say "assam" folder inside "data/map/transaction/hover/country/india/state"
    cur_state = path+state+"/" 
    map_year = os.listdir(cur_state)

    for year in map_year:
        # to access each year folder say "2018" folder inside "data/map/transaction/hover/country/india/state/assam/2018"
        cur_year = cur_state+year+"/"
        map_files = os.listdir(cur_year)

        for file in map_files:
            # to access each json file say "1.json" folder inside "data/map/transaction/hover/country/india/state/assam/2018/1.json"
            cur_file = cur_year+file
            with open(cur_file,"r") as f:
                info = json.load(f)

            for item in info['data']['hoverDataList']:
                map_transactions['District'].append(item['name'])
                map_transactions['Count'].append(item['metric'][0]['count'])
                map_transactions['Amount'].append(item['metric'][0]['amount'])
                map_transactions['State'].append(state)
                map_transactions['Year'].append(year)
                map_transactions['Quarter'].append(int(file.strip(".json")))

df_map_trans = pd.DataFrame(map_transactions)
```

Data Extraction for Map users

```python
path = data_dir+"/data/map/user/hover/country/india/state/"
map_state = os.listdir(path)

map_users = {"State": [], 
              "Year": [], 
              "Quarter": [], 
              "District": [],
              "RegisteredUser": [], 
              "AppOpens": []
            }


for state in map_state:
    # to access each state folder say "assam" folder inside "data/map/user/hover/country/india/state"
    cur_state = path+state+"/" 
    map_year = os.listdir(cur_state)

    for year in map_year:
        # to access each year folder say "2018" folder inside "data/map/transaction/hover/country/india/state/assam/2018"
        cur_year = cur_state+year+"/"
        map_files = os.listdir(cur_year)

        for file in map_files:
            # to access each json file say "1.json" folder inside "data/map/user/hover/country/india/state/assam/2018/1.json"
            cur_file = cur_year+file
            with open(cur_file,"r") as f:
                info = json.load(f)

            # print(info['data']['hoverData'].items())
            # dict_items
            # ([('north and middle andaman district', {'registeredUsers': 632, 'appOpens': 0}), 
            #   ('south andaman district', {'registeredUsers': 5846, 'appOpens': 0}), 
            #   ('nicobars district', {'registeredUsers': 262, 'appOpens': 0})
            # ])

            try:
                for item in info['data']['hoverData'].items():
                    map_users['District'].append(item[0])
                    map_users['RegisteredUser'].append(item[1]['registeredUsers'])
                    map_users['AppOpens'].append(item[1]['appOpens'])
                    map_users['State'].append(state)
                    map_users['Year'].append(year)
                    map_users['Quarter'].append(int(file.strip(".json")))
            except:
                pass

df_map_users = pd.DataFrame(map_users)
```

### Data Transformation

Establising connection to postgres database and creating tables

```python
conn = psycopg2.connect(database="phonepe_db",host="localhost",user="postgres",password="root",port="5432")
cursor = conn.cursor()

# aggregate_transactions table
cursor.execute('''CREATE TABLE IF NOT EXISTS agg_trans 
               (State varchar(100), Year int, Quarter int, Transaction_type varchar(100), Transaction_count int, Transaction_amount real)''')

# aggregate_users table  
cursor.execute('''CREATE TABLE IF NOT EXISTS agg_user 
               (State varchar(100), Year int, Quarter int, Brands varchar(100), Count int, Percentage real)''')

# map_transactions table
cursor.execute('''CREATE TABLE IF NOT EXISTS map_trans 
               (State varchar(100), Year int, Quarter int, District varchar(100), Count int, Amount real)''')

# map_users  table
cursor.execute('''CREATE TABLE IF NOT EXISTS map_user 
               (State varchar(100), Year int, Quarter int, District varchar(100), Registered_user int, App_opens int)''')

# top_transaction table
cursor.execute('''CREATE TABLE IF NOT EXISTS top_trans 
               (State varchar(100), Year int, Quarter int, Pincode int, Transaction_count int, Transaction_amount real)''')

# top_users table
cursor.execute('''CREATE TABLE IF NOT EXISTS top_user 
               (State varchar(100), Year int, Quarter int, Pincode int, Registered_users int)''')

conn.commit()
```
